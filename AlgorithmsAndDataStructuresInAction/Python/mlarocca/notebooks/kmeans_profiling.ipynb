```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import math
from typing import List
```

```python
KMEANS_STATS = '../tests/data/stats_kmeans.csv'
KMEANS_STATS_INC = '../tests/data/stats_kmeans_inc.csv'
KMEANS_STATS_BAK = '../tests/data/stats_kmeans.bak.csv'
```

```python
def plot_algo_stats(axe, df: pd.DataFrame, algorithm: str, by: str, time_field: str):
    df_algo: pd.DataFrame = df[df['algorithm']==algorithm]
    
    data = df_algo.sort_values(by=by)

    axe.plot(data[by], data[time_field], label=algorithm)
```

```python
def plot_k_stats(df: pd.DataFrame, method_name: str, k: int, time_field: str = 'cumulative_time'):
    df_k: pd.DataFrame = df[df['k']==k]
    algorithms = df_k['algorithm'].unique()

    fig, axe = plt.subplots()

    axe.set_title(f'{method_name}, k=={k}', fontsize=16)
    axe.set_xlabel('n')
    axe.set_ylabel('running time')
 
    for algorithm in algorithms:
        plot_algo_stats(axe, df_k, algorithm, 'n', time_field)

    axe.legend()
    plt.show()
```

```python
def plot_n_stats(df: pd.DataFrame, method_name: str, n: int, time_field: str = 'cumulative_time'):
    df_n: pd.DataFrame = df[df['n']==n][df['method_name'].str.startswith(method_name)]
    algorithms = df_n['algorithm'].unique()

    fig, axe = plt.subplots()

    axe.set_title(f'{method_name}, n=={n}', fontsize=16)
    axe.set_xlabel('k')
    axe.set_ylabel('running time')
 
    for algorithm in algorithms:
        plot_algo_stats(axe, df_n, algorithm, 'k', time_field)

    axe.legend()
    plt.show()
```

```python
def plot_method_stats(df: pd.DataFrame, method_name: str, ks: List[int], time_field: str = 'cumulative_time'):
    df_method: pd.DataFrame = df[df['method_name'].str.startswith(method_name)]
    for k in ks:
        plot_k_stats(df_method, method_name, k, time_field)
```
